## Role of weather in accidents

In the following section I have generated a new dataframe which will provide information regarding road accidents on the basis of weather.

Creating new columns which will store information about total road accidents, total deaths and injuries in road accidents in all the weathers of year 2016.

```python
total_accidents = weather_df['Total_accidents_in_fine_weather'] + + weather_df['Total_accidents_in_Hail'] + weather_df['Total_accidents_in_cloudy_weather']+ weather_df['Total_accidents_in_dust_storm']+ weather_df['Total_accidents_in_foggy_weather']+ weather_df['Total_accidents_in_other_weather']+ weather_df['Total_accidents_in_rainy_weather']+ weather_df['Total_accidents_in_snowfall']
```

```python
weather_df['Total_accidents_in_2016'] = total_accidents
```

```python
total_deaths = weather_df['Total_deaths_in_cloudy_weather'] + weather_df['Total_deaths_in_dust_storm'] + weather_df['Total_deaths_in_fine_weather'] + weather_df['Total_deaths_in_foggy_weather'] + weather_df['Total_deaths_in_other_weather'] + weather_df['Total_deaths_in_rainy_Hail'] + weather_df['Total_deaths_in_rainy_weather'] + weather_df['Total_deaths_in_snowfall']
```

```python
weather_df['Total_deaths_due_to_road_accidents'] = total_deaths
```

```python
total_injuries = weather_df['Total_injuries_in_cloudy_weather'] + weather_df['Total_injuries_in_dust_storm'] + weather_df['Total_injuries_in_fine_weather'] + weather_df['Total_injuries_in_foggy_weather'] + weather_df['Total_injuries_in_other_weather'] + weather_df['Total_injuries_in_rainy_Hail'] + weather_df['Total_injuries_in_rainy_weather'] + weather_df['Total_injuries_in_snowfall']
```

```python
weather_df['Total_injuries_due_to_road_accidents'] = total_injuries
```

```python
weather_df.columns 
```

```python
selected_columns = ['State_or_UT', 'Total_accidents_in_2016',  'Total_deaths_due_to_road_accidents', 'Total_injuries_due_to_road_accidents']
```

```python
all_weather_df = weather_df[selected_columns].copy()
```

```python
all_weather_df
```

```python
all_weather_df.describe()
```

You will notice by observing following two graphs that the pattern of number of accidents vary on the basis of season. As we can observe that state with maximum road accidents in fine weather is different from the state with maximum road accidents in rainy weather.

```python
plt.figure(figsize=(12,6))
plt.xticks(rotation=90)
plt.title("Road accidents in fine weather")
sns.barplot(y=weather_df.State_or_UT, x=weather_df.Total_accidents_in_fine_weather);
```

```python
plt.figure(figsize=(12,6))
plt.xticks(rotation=90)
plt.title("Road accidents in rainy weather")
sns.barplot(y=weather_df.State_or_UT, x=weather_df.Total_accidents_in_rainy_weather);
```

But it is always good to get the perspective from a large point of view. So I have plotted following graphs for total road accidents and total deaths in road accidents in year 2016.

```python
plt.figure(figsize=(12,6))
plt.xticks(rotation=90)
plt.title("Total road accidents in 2016")
sns.barplot(y=weather_df.State_or_UT, x=weather_df.Total_accidents_in_2016);
```

```python
plt.figure(figsize=(12, 6))
plt.title("Total deaths due to road accidents in 2016")
plt.xticks(rotation=90)
plt.plot(all_weather_df.State_or_UT, all_weather_df.Total_deaths_due_to_road_accidents,'o--r');
```
